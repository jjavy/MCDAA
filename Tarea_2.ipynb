## Crear, modificar o eliminar columnas

En este caso, para realizar un análisis más estádistico se procederá a hacer los siguientes cambios:

* Las columnas `Month_name` y `Weekday` son redundantes con su respectiva versión "sort", además al no ser valores numéricos, no merece la pena mantenerlos, por lo que serán **eliminados**.
* Los valores de `Time_of_Day` serán convertidos a valores; respectivamente: (Morning: 0, Afternoon: 1, Night: 2).
* Al igual que el caso anterior se le dará el valor 1 a "card" y 0 a "cash" en la columna `cash_type`.
* Siguiendo dicho patrón, `coffee_name` cuenta con solo 8 opciones, por lo que se asignará un valor a cada una:
  1. Americano
  2. Americano with Milk
  3. Latte
  4. Cappuccino
  5. Cortado
  6. Espresso
  7. Cocoa
  8. Hot Chocolate
* Finalmente se renombrará la columna `coffee_name` por un nombre que describa mejor su nuevo contenido.

```python
df_clnd = df.drop(columns=['Month_name', 'Weekday'])
df_clnd['Time_of_Day'] = df_clnd['Time_of_Day'].map({
      'Morning': 0,
      'Afternoon': 1,
      'Night': 2
    })
df_clnd['cash_type'] = df_clnd['cash_type'].map({'cash': 0, 'card': 1})
df_clnd['coffee_name'] = df_clnd['coffee_name'].map(
    {
        'Americano': 1,
        'Americano with Milk': 2,
        'Latte': 3,
        'Cappuccino': 4,
        'Cortado': 5,
        'Espresso': 6,
        'Cocoa': 7,
        'Hot Chocolate': 8
    }
)
df_clnd = df_clnd.rename(columns={'coffee_name': 'coffee_ordered'})
display(df_clnd)
```

## Filtrar registros
En este ejemplo se está filtrando las ventas superiores a 35 unidades monetarias (*no se especifica la moneda*) que hayan ocurrido en fin de semana.

```python
df_clnd.query('money > 35 and Weekdaysort >= 6')
```

## Crear un subconjunto de datos
En este ejemplo, se crea un subconjunto de todos los pedidos que son de bebidas de chocolate.

```python
chocolate_sales = df_clnd.query('coffee_ordered == 7 or coffee_ordered == 8')
display(chocolate_sales)
```

## Crear grupo de datos
En este ejemplo se crea un grupo de la media de dinero gastado en cada momento del día.

```python
grouped_data = df_clnd.groupby('Time_of_Day')['money'].mean()
display(grouped_data)
```
